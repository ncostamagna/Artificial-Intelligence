## Import the necessary software libraries:

```python
import numpy as np
import matplotlib.pyplot as plt

from envs import Maze
from utils import plot_policy, plot_action_values, test_agent
```

## Initialize the environment

```python
env = Maze()
```

```python
frame = env.render(mode='rgb_array')
plt.axis('off')
plt.imshow(frame)
```

## Define value table $Q(s, a)$

#### Create the $Q(s, a)$ table

```python
action_values = np.zeros(shape=(5, 5, 4))
```

#### Plot $Q(s, a)$

```python
# cada cuadrado representa un estado
# cada calor de cada cuadrado representa una accion
plot_action_values(action_values)
```

## Define the policy $\pi(s)$

#### Create the policy $\pi(s)$

```python
# Lo hacemos estocastico
def policy(state, epsilon=0.2):
    if np.random.random() < epsilon:
        return np.random.randint(4) # escojera al azar una de las 4 acciones
    else:
        av = action_values[state]
        return np.random.choice(np.flatnonzero(av == av.max()))  # escojera el maximo
        # Si hay mas de un maximo igual, devolvemos cualquiera
```

#### Test the policy with state (0, 0)

```python
action = policy((0,0))
print(f"Action taken in state (0,0): {action}")
```

#### Plot the policy

```python
plot_policy(action_values, frame)
```

## Implement the algorithm


</br></br></br>




<div style="text-align:center">
    Adapted from Barto & Sutton: "Reinforcement Learning: An Introduction".
</div>

```python
def on_policy_me_control(policy, action_values, episodes, gamma=0.99, epsilon=0.2):
    
    sa_returns = {} # registros de los retornos cada vez que tomamos una desicion
    
    for episode in range(1, episodes + 1):
        state = env.reset()
        done = False # se termino la tarea o no
        transitions = [] # transicion de estados, la accion que tomamos y la recompensa que obtuvimos
        plot_action_values(action_values)
        
        while not done:
            action = policy(state, epsilon)
            next_state, reward, done, _ = env.step(action)
            transitions.append([state, action, reward]) # guardamos la transicion
            state = next_state # accionamos el nuevo estado
            
        
        G = 0 #retorno
        
        # iteramos por las transiciones guardadas, desde la ultima hasta la primera
        for state_t, action_t, reward_t in reversed(transitions):
            G = reward_t + gamma * G 
            
            if not (state_t, action_t) in sa_returns:
                sa_returns[(state_t, action_t)] = []
            
            sa_returns[(state_t, action_t)].append(G) # guardamos el retorno G
            #print(sa_returns)
            action_values[state_t][action_t] = np.mean(sa_returns[(state_t, action_t)]) # media para ese par estado accion
```

```python
on_policy_me_control(policy, action_values, episodes=5)
```
